### Visualize `Adj Close` for each company as it is

### Create a new `pd.DataFrame`, `adj_close_df` which will contain the `Adj Close` values from the dataframes of all the companies in `df_dict`

```python
adj_close_df = pd.DataFrame()

for company in df_dict.keys():
    adj_close = df_dict[company]["Adj Close"].copy(deep = True)
    t = df_dict[company]["Date"].copy(deep = True)
    t = pd.to_datetime(t, format = "%Y-%m-%d")
    adj_close.index = t
    adj_close_df[company] = adj_close
```

### Perform delay to compute Expected Return



- _Expected Return_ measures the mean, or expected value, of the probability distribution of investment returns. 
- The _Expected Return_ of a portfolio is calculated by multiplying the weight of each asset by its expected return and adding the values for each investment.

Formula used: $r_t = \frac{p_t - p_{t - 1}}{p_{t - 1}} = \frac{p_t}{p_{t - 1}} - 1$

```python
plt.rcParams['figure.figsize'] = 18, 12

for i, company in enumerate(moving_averaged.columns):
    r = moving_averaged[company].copy(deep = True)
    r = (r/r.shift(1)) - 1
    t = r.index
    r.reset_index(drop = True, inplace = True)
    plt.subplot(n, 1, i + 1)
    plt.plot(t, r)
    plt.title(company + "_Expected_Return")

plt.tight_layout()
plt.show()
```

## Price Prediction by Forecasting

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

from sklearn.neighbors import KNeighborsRegressor

from sklearn.pipeline import make_pipeline
```

### Forecast Class Pipeline

```python
from collections import defaultdict
from copy import deepcopy

class forecast:
    def __init__(self, X, Y):
        self.X = deepcopy(X)
        self.Y = deepcopy(Y)
        self.models = defaultdict(lambda: None)
        self.scaler = None
    def train_test_split(ratio):
        n = X.shape[0]
        n_train = n * ratio
        self.X_train = self.X[:n_train, :]
        self.Y_train = self.Y[:n_train, :]
        self.X_test = self.X[n_train:, :]
        self.Y_test = self.Y[n_train:, :]
    def set_feature_scaler(X_scaler, Y_scaler):
        self.X_scaler = X_scaler
        self.Y_scaler = Y_scaler
    def scale_training_data():
        if self.X_scaler is not None and self.Y_scaler is not None:
            self.X_train_scaled = X_scaler.fit_transform(self.X_train)
            self.Y_train_scaled = Y_scaler.fit_transform(self.Y_train)
    def scale_testing_data():
        if self.X_scaler is not None and self.Y_scaler is not None:
            self.X_test_scaled = X_scaler.transform(self.X_test)
            self.Y_test_scaled = Y_scaler.transform(self.Y_test)
    def add_ML_model(self, model_name, sklearn_model_object):
        self.models[model_name] = sklearn_model_object
    def predict_using_model(self, model_name):
        try:
            Y_hat = self.models[model_name].predict(self.X_test_scaled)
            Y_hat = self.Y_scaler.inverse_transform(Y_hat)
            return(Y_hat)
        except:
            return(None)
```
